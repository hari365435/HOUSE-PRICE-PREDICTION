```python
import numpy as np
import joblib
import pandas as pd
data=pd.read_csv("House_Price_Data.csv")
print(data)
df=data

```

```python

tier1=['Ahmedabad','Amritsar','Bangalore','Bhilai', 'Bhopal', 'Bhubaneswar',
 'Bilaspur' ,'Chennai' ,'Coimbatore' ,'Cuttack' ,'Dehradun' ,'Dhanbad']
tier2=['Durgapur' ,'Dwarka' ,'Ernakulam', 'Faridabad', 'Gurgaon' ,'Guwahati',
 'Gwalior', 'Haridwar' ,'Howrah' ,'Hyderabad', 'Indore', 'Jaipur', 'Jalandhar']

print('Ahmedabad' in tier1)
```

```python

def city_tier(loc):
    if loc in tier1:
        return 1
    elif loc in tier2:
        return 2
    else:
        return 3
        
df["city_tier"]=df["location"].apply(city_tier)
print(df)    
```

```python
print(df["propertytype"].unique())
```

```python
def pp(type):
    if type =="Flat":
        return 1
    elif type=="Villa":
        return 2
    else:
        return 3
df["pp"]=df["propertytype"].apply(pp)
```

```python
print(df.head(-1))
```

```python
x=df[["bhk","pp","city_tier","sqft"]]
y=df["totalprice"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=21,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(y_pred)
print(r2_score(y_test,y_pred))
      
```

```python

joblib.dump(model, "ml_model.pkl")
print("Model saved as ml_model.pkl")
```
